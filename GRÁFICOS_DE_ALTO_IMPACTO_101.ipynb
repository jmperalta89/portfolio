#VERSIÓN EXPRESS

Importación de librerías

```python
import pandas as pd
import plotly.express as px
```

Preparación de los datos


```python
# Datos extendidos hasta 2024
data = {
    'Country': ['USA', 'China', 'Japan', 'Germany', 'UK'] * 10,
    'Year': [2015, 2015, 2015, 2015, 2015,
             2016, 2016, 2016, 2016, 2016,
             2017, 2017, 2017, 2017, 2017,
             2018, 2018, 2018, 2018, 2018,
             2019, 2019, 2019, 2019, 2019,
             2020, 2020, 2020, 2020, 2020,
             2021, 2021, 2021, 2021, 2021,
             2022, 2022, 2022, 2022, 2022,
             2023, 2023, 2023, 2023, 2023,
             2024, 2024, 2024, 2024, 2024],
    'GDP': [18200, 11000, 4400, 3400, 2900,
            18700, 11200, 4500, 3500, 3000,
            19500, 12200, 4800, 3700, 3100,
            20500, 13600, 5000, 3900, 3200,
            21400, 14300, 5100, 4000, 3300,
            20900, 14700, 5000, 3800, 2700,
            23000, 17700, 4900, 4200, 3200,
            25000, 18000, 4200, 4300, 3400,
            27000, 19000, 4400, 4500, 3600,
            28500, 20000, 4600, 4700, 3800]
}

df = pd.DataFrame(data)
```

Creación y visualización del gráfico

```python
fig = px.bar(df, x="GDP", y="Country", animation_frame="Year",
             orientation='h', range_x=[0,40000],
             color="Country", hover_name="Country",
             title="GDP por País a lo Largo del Tiempo")

fig.update_layout(xaxis_title="GDP (Miles de Millones USD)", yaxis_title="País")
fig.show()
```

Explicación

```python
fig = px.bar(
    df,                       # DataFrame que contiene los datos
    x="GDP",                  # Columna del DataFrame que se usará para el eje x
    y="Country",              # Columna del DataFrame que se usará para el eje y
    animation_frame="Year",   # Columna del DataFrame que se usará para la animación en función del tiempo
    orientation='h',          # Orientación del gráfico, en este caso 'h' indica que será un gráfico de barras horizontal
    range_x=[0, 40000],       # Rango del eje x, en este caso de 0 a 40,000
    color="Country",          # Columna del DataFrame que se usará para colorear las barras
    hover_name="Country",     # Columna del DataFrame que se mostrará al pasar el cursor sobre una barra
    title="GDP por País a lo Largo del Tiempo"  # Título del gráfico
)
```

Importo librerías

```python
import bar_chart_race as bcr
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from random import randint
import bar_chart_race as bcr
```

Cargo dataset

```python
df_2 = bcr.load_dataset('covid19_tutorial')
```

Configuro fuente

```python
# Configuración global de Matplotlib
plt.rcParams['font.family'] = 'DejaVu Sans'
```

Genero el gráfico

```python
bcr.bar_chart_race(
    df=df_2,
    filename='covid19_horiz.mp4',
    orientation='h',
    sort='desc',
    n_bars=6,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=10,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_fmt='%B %d, %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Total deaths: {v.nlargest(6).sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'DejaVu Sans'},
    perpendicular_bar_func='median',
    period_length=500,
    figsize=(5, 3),
    dpi=144,
    cmap='dark12',
    title='COVID-19 Deaths by Country',
    title_size=10,
    bar_label_size=7,
    tick_label_size=7,
    shared_fontdict={'family' : 'DejaVu Sans', 'color' : '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=False)
```
